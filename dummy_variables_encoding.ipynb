## Method 1: using pd.get_dummies()

```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df_org=pd.read_csv('carprices.csv')
df_org
```

```python
plt.scatter(df_org[['Mileage']],df_org['Sell Price($)'])
```

```python
dum=pd.get_dummies(df_org['Car Model'])
dum
```

```python
df_joint=pd.concat([df_org,dum],axis='columns')
df_joint
```

```python
df_joint=df_joint.drop('Car Model',axis='columns')
```

```python
df_joint
```

```python
df_joint=df_joint.drop('Mercedez Benz C class',axis='columns')
df_joint
```

```python
X=df_joint.drop('Sell Price($)',axis='columns').values
X
```

```python
Y=df_joint['Sell Price($)']
Y
```

```python
car_mod=linear_model.LinearRegression()
car_mod.fit(X,Y)
```

```python
car_mod.predict([[45000,4,0,0]])   # benz , 4 yrs old and mileage 45000
```

```python
car_mod.predict([[86000,7,0,1]])   # BMW , 7 yrs ,  86000 
```

```python
car_mod.score(X,Y)
```

## method 2 : using sklearn.preprocessing  (LabelEncoder,OneHotEncoder)

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lab_en=LabelEncoder()
```

```python
df_main=pd.read_csv('carprices.csv')
df_main
```

```python
df_main['Car Model']=lab_en.fit_transform(df_main['Car Model'])
```

```python
df_main
```

```python
X=df_main.drop('Sell Price($)',axis='columns').values
X
```

```python
Y=df_main['Sell Price($)']
Y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('Car Model', OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python

X = ct.fit_transform(X)
X
```

```python
X=X[:,1:]
X
```

```python
mod_car=linear_model.LinearRegression()
mod_car.fit(X,Y)
```

```python
mod_car.predict([[0,1,45000,4]])
```

```python
mod_car.predict([[1,0,86000,7]]) 
```
